```python
def factorial(n):
  fact = 1
  for i in range(1,n+1):
    fact = fact*i
  return fact

factorial(10)
```

```python
def binary_numbers(bn):
  if bn == 1: return 1

  remainder_list = []

  while bn > 0:
      remainder = bn % 2
      bn = bn//2
      remainder_list.append(remainder)

  count = 0
  max_ = 0

  for i in remainder_list:
      if i == 1:
          count += 1
          if max_ < count:
              max_ = count
      else:
          count = 0

  return max_

binary_numbers(125)

```
